```python
import io
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from copy import deepcopy
from google.colab import files
import plotly.express as px
```

```python
uploaded =files.upload()
```

```python
strikers_df=pd.read_csv(io.BytesIO(uploaded["top_fifty_striker.csv"]))
strikers_df
```

```python
strikers_df_radar=deepcopy(strikers_df)
```

```python
strikers_df_radar[strikers_df_radar['player_name'].str.contains("Harry")]
```

```python
strikers_df_radar.columns
```

```python
px.bar(x='player_name',y='time',text='time', data_frame=strikers_df_radar)
```

```python
per90cols=['goals','assists','xG','shots','key_passes','xA','xGChain','xGBuildup']
for col in per90cols:
  strikers_df_radar[col+"Per90"]=strikers_df_radar[col].divide(strikers_df_radar['time']).multiply(90)
```

```python
cols_radar=[i+"Per90" for i in per90cols]
cols_radar
```

```python
strikers_df_radar.loc[(strikers_df_radar['player_name']=="Harry Kane"),cols_radar].sum()
```

```python
fig=px.line_polar(strikers_df_radar,r=strikers_df_radar.loc[(strikers_df_radar['player_name']=="Harry Kane"),cols_radar].sum(),theta=cols_radar,line_close=True)
fig.update_traces(fill='toself')
fig.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
strikers_df_radar[cols_radar]=scaler.fit_transform(strikers_df_radar[cols_radar])
```

```python
fig=px.line_polar(strikers_df_radar,r=strikers_df_radar.loc[(strikers_df_radar['player_name']=="Harry Kane"),cols_radar].sum(),theta=cols_radar,line_close=True)
fig.update_traces(fill='toself')
fig.show()
```

```python
strikers_df_radar.loc[(strikers_df_radar['player_name']=="Harry Kane"),cols_radar].sum()
```

```python
radar_max=strikers_df_radar[cols_radar].max().max()
radar_max
```

```python
import plotly.graph_objects as go
```

```python
for i, row in strikers_df_radar.iterrows():
  if row["id"]==647:
    continue
  print(row["player_name"])
  fig=go.Figure()
  fig.add_trace(go.Scatterpolar(
        r=strikers_df_radar.loc[(strikers_df_radar['id']==647),cols_radar].sum(),theta=cols_radar,
fill='toself',name="Harry Kane"))
  fig.add_trace(go.Scatterpolar(
        r=strikers_df_radar.loc[(strikers_df_radar['id']==row['id']),cols_radar].sum(),theta=cols_radar,
fill='toself',name=row["player_name"]))
  fig.update_layout(title="Harry Kane vs " + row["player_name"] + " of " +row["team_title"],
    polar=dict(radialaxis=dict(visible=True,range=[0,radar_max])),
    showlegend=True)
  fig.show()
```
